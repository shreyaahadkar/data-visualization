```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips
```

## Scatter Plot

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette='summer')
```

## Histogram

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

## Bar Plot

```python
sns.barplot(x='sex', y='tip', data=tips, palette='coolwarm')
```

## Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='YlGnBu')
```

## Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', palette='spring', dodge=True)
```

## Joint Plot

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='icefire')
```

## Pair Plot

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```

## Heat Map

```python
sns.heatmap(titanic.corr(), annot=True, cmap='YlGnBu')
```

## Cluster Map

```python
sns.clustermap(iris.drop('species', axis=1))
```
