```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/heart-failure-prediction/heart.csv')
```

```python
df.head()
```

## Distribution Plot for Numeric Columns

```python
sns.displot(df['Age'], kde='True', rug=True)
```

```python
sns.displot(df['RestingBP'], kde=True)

# avg resting bp is bw 125-150
```

```python
sns.displot(df['Cholesterol'], kde=True)
```

```python
sns.displot(df['MaxHR'], kde=True)
```

## Pie Chart for Categorical Columns

```python
df.groupby('ChestPainType').size().plot(kind='pie', autopct='%.1f%%')
```

```python
df.groupby('FastingBS').size().plot(kind='pie', autopct='%.1f%%')
```

```python
df.groupby('RestingECG').size().plot(kind='pie', autopct='%.1f%%')
```

```python
df.groupby('ST_Slope').size().plot(kind='pie', autopct='%.1f%%')
```

```python
df.groupby('HeartDisease').size().plot(kind='pie', autopct='%.1f%%')
```
