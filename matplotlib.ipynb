```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,4*np.pi,100)

sin_x = np.sin(x)
cos_x = np.cos(x)

plt.plot(x, sin_x, label = 'Sin', color = 'red')
plt.title('Sine Wave')

plt.plot(x, cos_x, label = 'Cos', color = 'green')
plt.title('Sine and Cosine Wave')

plt.legend()
```

```python
x = np.linspace(0,4*np.pi,100)

sin_x = np.sin(x)
cos_x = np.cos(x)

# Using markers
plt.plot(x, sin_x,'*')
plt.title('Sine Wave')

plt.plot(x, cos_x,'--')
plt.title('Sine and Cosine Wave')
```

```python
#scatter plot
plt.style.use('Solarize_Light2')

x = np.random.rand(1000)
y = np.random.randn(1000)

x1 = np.random.rand(1000) + 1
y1 = np.random.randn(1000) + 1

plt.figure(figsize = (10,5))
plt.title('Scatter Plot')

plt.scatter(x,y,color = 'red', alpha = .8, s = 10, label = 'Group-1')
plt.scatter(x1,y1,color = 'blue', alpha = .8, s = 10, label = 'Group-2')

plt.legend()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

plt.figure(figsize = (7,7))

x = iris.data

plt.scatter(x[:,0], x[:,1], s = 50, marker = 'x')

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
```

```python
plt.style.use('grayscale')
x = np.random.randn(100)
y = np.random.randn(100)

colors = np.random.randn(100)
size = np.random.randn(100) * 1000

plt.title("Random Scatter Plot")
plt.xlabel('x values')
plt.ylabel('y values')

plt.scatter(x,y, c = colors, s = size, alpha = .7, cmap = 'viridis')

plt.colorbar()
```

```python
plt.style.use('classic')
data = np.random.randn(1000)
data2 = np.random.randn(1000) + 2
plt.xlabel('values')
plt.ylabel('freq')
plt.hist(data, color = 'red', bins = 20, alpha = .8, edgecolor = 'black')
plt.hist(data2, color = 'blue', bins = 20, alpha = .8, edgecolor = 'black')
plt.title('Histogram')
plt.show()
```
